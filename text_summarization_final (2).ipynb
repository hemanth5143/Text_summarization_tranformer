```python
import pandas as pd
```

```python
data = pd.read_excel('/content/Inshorts Cleaned Data.xlsx')
data
```

```python
data = data.rename(columns={"Headline":"target_text", "Short":"source_text"})
data = data[['source_text', 'target_text']]
```

```python
data
```

```python
data['source_text'] = "summarize: " + data['source_text']
data
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(data, test_size=0.1)
train_df.shape, test_df.shape
```

```python
! pip install simplet5 -q
```

```python
from simplet5 import SimpleT5

model = SimpleT5()
model.from_pretrained(model_type="t5", model_name="t5-base")
```

```python
model.train(train_df=train_df[:],
            eval_df=test_df[:1000], 
            source_max_token_len=128, 
            target_max_token_len=50, 
            batch_size=8, max_epochs=2, use_gpu=True)
```
